```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/coaching_student_data.csv")

df.head()
```

```python
df.info()
```

```python
#Clean column names

df.columns = df.columns.str.strip()
```

```python
#Total Marks of students
df["Total"]=df.iloc[:,1:7].sum(axis=1)
```

```python
#Percentage of students

df["Percentage"] = (df["Total"] / 600)*100
```

```python
#Add Grade

def assign_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 75:
        return "A"
    elif percentage >= 60:
        return "B"
    elif percentage >= 50:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "Fail"

df["Grade"] = df["Percentage"].apply(assign_grade)
```

```python
#Saved cleaned dataset

df.to_csv("../data/coaching_student_cleaned.csv", index=False)
```

```python
#Top Students

top_students = df.sort_values(by="Percentage", ascending=False).head(5)

top_students[["Name", "Percentage"]]
```

```python
#Weak students 

weak_students = df[df["Percentage"] < 40]
if weak_students.empty:
    print("No weak student in the class !")
else:
    print("Weak students")
    weak_students[["Name", "Percentage"]]
```

```python
#To fins weak subject of a studenst
subjects=["English", "Hindi", "Maths", "Science", "Social Science", "Computer"]
def find_weak_subjects(row):
    weak=[]
    for subject in subjects:
        if row[subject] < 40:
            weak.append(subject)
    if len (weak) == 0:
        return "No particular weak subject"
    return ", ".join(weak)
```

```python
#Applying to dataset

df["Weak Subjects"] = df.apply(find_weak_subjects, axis=1)
```

```python
#Subject wise Average marks in the class

subject_avg = df.iloc[:, 1:7].mean()

subject_avg
```

```python
#Hardest subject

hardest_subject = subject_avg.idxmin()
```

```python
#Teacher Insight

print("📌 Teacher Insight:")
print(f"Students are weak in {hardest_subject}")
```

```python
#Students Needing Extra Attention

attention_students = df[df["Percentage"] < 60]

attention_students[["Name", "Percentage","Weak Subjects"]]
```

```python
#Class Performance

print("📊 Class Summary")

print("Average Percentage of class:", df["Percentage"].mean())
print("Highest Score of class:", df["Percentage"].max())
print("Lowest Score of class:", df["Percentage"].min())
```

```python
# Subject Weakness


subjects = ["English", "Hindi", "Maths", "Science", "Social Science", "Computer"]

weak_counts = {}

for sub in subjects:
    weak_counts[sub] = (df[sub] < 40).sum()

for sub , count in weak_counts.items():
    print(f"Weak in {sub}: {count} students ")
```

```python
# Subject-wise Performance

def categorize_student(percentage):
    if percentage >= 75:
        return "Good"
    elif percentage >= 50:
        return "Average"
    else:
        return "Weak"

df["Category"] = df["Percentage"].apply(categorize_student)

df[["Name", "Percentage", "Category"]]
```

```python
#Weak Subject of students

subjects = ["English", "Hindi", "Maths", "Science", "Social Science", "Computer"]

def get_weak_subjects(row):
    weak = []
    for sub in subjects:
        if row[sub] < 40:
            weak.append(sub)
        if len(weak) == 0:
            return "No weak subject"
    return ", ".join(weak)
    
df["Weak Subjects"] = df.apply(get_weak_subjects, axis=1)

df[["Name", "Weak Subjects"]]
```

```python
#Strong subject of students

def get_strong_subjects(row):
    strong = []
    for sub in subjects:
        if row[sub] >= 75:
            strong.append(sub)
        if len(strong) == 0:
            return "No particular strong subject"
    return ", ".join(strong)

df["Strong Subjects"] = df.apply(get_strong_subjects, axis=1)

df[["Name", "Strong Subjects"]]
```

```python
#Student Performance 

def performance_type(row):
    if row["Category"] == "Strong" and row["Weak Subjects"] != "":
        return "Strong but Weak in Some Subjects"
    elif row["Category"] == "Weak" and row["Strong Subjects"] != "":
        return "Weak but Good in Some Subjects"
    else:
        return "Consistent Performance"

df["Performance Type"] = df.apply(performance_type, axis=1)

df[["Name", "Category", "Performance Type"]]
```

```python
df.to_csv("../data/coaching_student_final.csv", index=False)
```

```python
#Adding Recommendations

def generate_recommendation(row):
    
    weak = row["Weak Subjects"]
    strong = row["Strong Subjects"]

    # STRONG STUDENTS
    if row["Category"] == "Strong":
        if weak and weak != "No weak subject":
            return f"Strong overall, but needs improvement in {weak}"
        else:
            return "Excellent performance. Keep it up!"
    
    # AVERAGE STUDENTS
    elif row["Category"] == "Average":
        if weak and weak != "No weak subject":
            return f"Can improve. Focus on {weak}"
        else:
            return "Good performance. Try to reach strong category."
    
    # WEAK STUDENTS
    else:
        if strong and strong != "":
            return f"Needs improvement overall, but doing well in {strong}"
        else:
            return "Needs serious improvement in all subjects"
```

```python
#Applying recommendations to dataset

df["Recommendation"] = df.apply(generate_recommendation, axis=1)

df[["Name", "Recommendation"]]
```

```python
df.to_csv("../data/coaching_student_final.csv", index=False)
```
